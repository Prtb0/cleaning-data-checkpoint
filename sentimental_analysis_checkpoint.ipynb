```python
# setting up stop words for sentimental analysis
nlp = spacy.load("en_core_web_sm")
sw_spacy = nlp.Defaults.stop_words
additional_list = ['airline','aircraft','plane','flight','airport']
sw_spacy.update(additional_list)
print(sw_spacy)
```

```python
# Read positive reviews airlines data
df = pd.read_csv("C:\\Users\\prave\\Desktop\\LaunchcodeW+\\cleaning-data-checkpoint\\Airline_review_cleaning_checkpoint.csv")
pos_sentiment_df= df[(df['Airline Name']== 'Bangkok Airways' ) | (df['Airline Name']== 'Hainan Airlines') |
                    (df['Airline Name'] == 'China Southern Airlines') | (df['Airline Name'] == 'QantasLink') |
                    (df['Airline Name'] == 'ANA All Nippon Airways')]
```

```python
# Constructing  nltk stop words
for index, row in pos_sentiment_df.iterrows():
    doc = nlp(row['Review_Title'])
#   unwanted words adding to Spacy stop words to minimize noise in word cloud
    sw_spacy.update([token.lemma_ for token in doc if token.pos_ == "VERB"])
```

```python
# Splitting sentences into words
pos_words_list = []
for index, row in pos_sentiment_df.iterrows():
    rw = TextBlob(row['Review_Title'])
    pos_words_list.extend(list(rw.words))
```

```python
# making final positive words after filtering out stop words
pos_review_words = [word for word in pos_words_list if word.lower() != '0' and word.lower() not in sw_spacy]
print(pos_review_words)
```

```python
# New dataframe column to hold words from review text
pos_sentiment_df['cleaned_review_words'] = pd.Series(pos_review_words)
```

```python
# Read the data for sentiment analysis
sentiment_df = pd.read_csv("C:\\Users\\prave\\Desktop\\LaunchcodeW+\\cleaning-data-checkpoint\\Airline_review_cleaning_checkpoint.csv")
```

```python
# Create a function to get the subjectivity
def getSubjectivity(review):
   return TextBlob(review).sentiment.subjectivity

#Create a function to get the polarity
def getPolarity(review):
    return TextBlob(review).sentiment.polarity

  
 #Create two new columns ‘Subjectivity’ & ‘Polarity’
sentiment_df['Review_Subjectivity'] = sentiment_df['Review'].apply(getSubjectivity)
sentiment_df ['Review_Polarity'] = sentiment_df['Review'].apply(getPolarity)
 
def getAnalysis(score):
    if score < 0:
        return '-1'
    elif score == 0:
        return '0'
    else:
        return '1'
    

# Calculatig final sentiment score 
sentiment_df['Review_Sentiment_Score'] = sentiment_df['Review_Polarity'].apply(getAnalysis )
```

```python
# DF after calculating sentiment polarity and subjectivity
sentiment_df[['Review_Title','Review_Subjectivity','Review_Polarity',"Review",'Review_Sentiment_Score']]
```
